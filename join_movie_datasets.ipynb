```python
import pandas as pd
```

```python
movies = pd.read_csv("title.movies.tsv", sep='\t'
                                        , dtype={'tconst':'string'
                                            , 'titleType':'string'
                                            , 'primaryTitle':'string'
                                            , 'originalTitle':'string'
                                            , 'isAdult':'float64'
                                            , 'startYear':'float64'
                                            , 'endYear':'float64'
                                            , 'runtimeMinutes':'float64'
                                            , 'genres':'string'
                                            }
                                            , na_values='\\N'
                                                          )
print(movies.shape)
movies.head(3)
```

```python
recent_movies = movies[movies['startYear']>=1980][['tconst','primaryTitle','startYear']].copy()
recent_movies.shape
```

```python
actor_movies = pd.read_csv("title.actors.tsv", sep='\t')
actor_movies.head(3)
```

```python
ratings = pd.read_csv("title.ratings.tsv", sep='\t')
ratings.head(3)
```

```python
ratings_above_6 = ratings[ratings['averageRating']>6].copy()
```

```python
person_names = pd.read_csv("name.basics.tsv", sep='\t')
person_names.head(3)
```

```python
us_titles = pd.read_csv("title.us.tsv", sep='\t').rename(columns={'titleId':'tconst'})
us_titles_filtered = us_titles[(us_titles['region']=='US')
                               ].copy()
us_titles_filtered.head(3)
```

```python
# us_titles_filtered[us_titles_filtered['title']=='']
```

```python
actor_movies_ratings = (actor_movies
                        .merge(ratings_above_6, how='inner', on='tconst')
                        .merge(recent_movies, how='inner', on='tconst')
                        .merge(us_titles_filtered[['tconst','region','language']], how='inner', on='tconst')
                        .merge(person_names[['nconst','primaryName','birthYear']],how='inner', on='nconst')
                        )
actor_movies_ratings.head(3)
```
